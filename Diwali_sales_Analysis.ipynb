## Exploratory Data Analysis

```python
ax = sn.countplot(x = 'Gender',data = df) ## here the values are not coming on bar . see next step
```

## Age

```python
df.columns
```

```python
sn.countplot(data=df,x="Age Group")  ## This is without hue
```

```python
ax = sn.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax = sn.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Total Amount vs Age group
```

```python
sales = df.groupby(['Age Group'],as_index=False).sum(['Amount']).sort_values(by='Amount',ascending=False)
sn.barplot(x='Age Group',y='Amount',data = sales);

```

 From the above graphs we can see that most of the buyers are of age group 26-35

## State

```python
df.columns
```

```python
state_sales = df.groupby(['State'],as_index=False).sum(['Orders']).sort_values(by='Orders',ascending=False).head(10)
sn.set(rc={'figure.figsize':(18,5)})
axx= sn.barplot(data=state_sales,x='State',y='Orders')

```

 so here the breakdown of the code
 1) First we groupby states and how many orders per state and head for top 10
    then we set the figure of barplot by putting set and rc=: 
 2) This parameter stands for "runtime configuration" (or "run configuration") and is used to           specify default settings     for 
    visualizations. rc accepts a dictionary where keys are configuration settings and values are  the desired settings.
 3) And then the barplot 
 4) if we dont set the figure then the names of the states will overpal and we wont understand anything

From the above graph we can see that most of the orders and sale/amount are from UttarPradesh,Maharashtra,Karnataka respectively

## Marital Status

```python
df.columns
```

```python
ax = sn.countplot(x= 'Shadi',data= df)

sn.set(rc={'figure.figsize':(7,5)});
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Shadi','Gender'],as_index=False).sum('Amount').sort_values(by='Amount',ascending=False)
sn.set(rc={'figure.figsize':(6,5)})
sn.barplot(data=sales_state,x='Shadi',y='Amount',hue='Gender')
```

From the above graph we can see that most of the buyers are married women 

## Occupation

```python
df.columns
```

```python
sn.set(rc={'figure.figsize':(20,5)})
ax = sn.countplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_by_occupation = df.groupby(['Occupation'],as_index=False).sum(['Amount']).sort_values(by='Amount',ascending=False)
sn.set(rc={'figure.figsize':(20,5)})
sn.barplot(data=sales_by_occupation,x='Occupation',y='Amount');
```

From above graphs we can see that most of the buyers are working in IT,Healthcare and Aviation sector

## Product Category

```python
sn.set(rc={'figure.figsize':(24,5)})
ax = sn.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_sales = df.groupby(['Product_Category'],as_index=False).sum(['Amount']).sort_values(by='Amount',ascending=False).head(10)
sn.set(rc={'figure.figsize':(20,5)})
sn.barplot(data=product_sales,x='Product_Category',y='Amount')
```

From the above graph we can say that Food , Clothing & Apparel , Electronics & Gadgets are the top3 products sales wise
